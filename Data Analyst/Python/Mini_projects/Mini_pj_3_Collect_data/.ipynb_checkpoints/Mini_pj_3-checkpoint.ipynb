````python
# красивый вариант объединения df's to main DF
# 'current_path' - название пути; 'dirs' - список из названий всех папок; 'files' - список из названий всех файлов


### Creat general DF (main DF) for all data  
df = pd.DataFrame()              # созданный пустой DF в который в дальнейшем будут попадать файлы из папок, DF-склейка


### Read data from all files and add it to general df
for current_path, dirs, files in os.walk(path):      # красивый вариант обращения к переменным в цикле
    for file in files:
        ### Contruct full path to file
        data_path = f'{current_path}/{file}'      # аналог записи: current_path +'/'+ file
                                                  # аналог: os.path.join('...', '....', ...) - соединения пути с авто подставлением нужного сепоратора

        ### Read data
        temp_df = pd.read_csv(data_path)             # read new df 

        ### Extract date and name
        path_parts = current_path.split('\\')        # добавление колонок name, date в general DF из пути
        date = path_parts[-2]
        name = path_parts[-1]
        
        ### Add columns 'date', 'name' to new df (to temp_df) whith data
        temp_df['date'] = date
        temp_df['name'] = name

        ### Rebuild general DF whith all data (Insert data into main DF)
        df = pd.concat((df, temp_df))                # перезапись generel DF склейкой с new DF  
        
print(df.head())                                    # print перезаписанного general DF со всеми файлами в указанном пути 
````

````python
df.drop(columns='Unnamed: 0', inplace= True)   # drop unnecessary column
````

````python
df.reset_index(drop = True, inplace= True)                # drop old index to redo index to normal values
````

````python
### Creat general DF (main DF) for all data  
df = pd.DataFrame()             

### Read data from all files and add it to general df
for current_path, dirs, files in os.walk(path):     
    
    ### Contruct full path to file
    for file in files: 
        part_path = Path(current_path)
        data_path = part_path / file        # Contruct full path to file
    
        ### Read data
        temp_df = pd.read_csv(data_path)             
        
        ### Extract date and name
        name = data_path.parts[-2]          # аналог: data_path.name - выводит последний элемент пути - имя файла
        date = data_path.parts[-3]          # data_path.parts - кортеж со всеми компонентами пути, которые не нужно сплитить как в прошлом варианте
        
        ### Add columns 'date', 'name' to new df (to temp_df) whith data
        temp_df['date'] = date
        temp_df['name'] = name

        ### Rebuild general DF whith all data (Insert data into main DF)
        df = pd.concat((df, temp_df))                
        
print(df.head())                                    
````

````python
# Create sort df by sum(quantity)
sort_quantity_df = df\
                    .groupby('name')\
                    .agg({'quantity':'sum'})\
                    .sort_values('quantity', ascending = False)

sort_quantity_df
````

````python
# Create sort df by sum(quantity)
sort_quantity_df = df\
                    .groupby('name')\
                    .agg({'quantity':'sum'})\
                    .sort_values('quantity', ascending = False)

# Finde max quantity orders
max_quantity_orders = sort_quantity_df.quantity.max()

# Create list whith top clients, who orders max quantity produts (names in this list have wrong form: "Name_Surname")
default_list_top_customers = sort_quantity_df\
                                .query('quantity == @max_quantity_orders')\
                                .index
default_list_top_customers = list(default_list_top_customers)

# Rebuild list whith top customers into correct form: "Surname Name" 
list_top_customers = []
for name in default_list_top_customers:
    cut_name = name.split('_')
    full_name = '{} {}'.format(cut_name[-1],cut_name[0])
    list_top_customers.append(full_name)

# Ascending names in list   
list_top_customers.sort()    

print('''Пользователи, купившие максимальное количество товаров: {}. 
Количество товаров: {}.'''.format(', '.join(map(str, list_top_customers)),  max_quantity_orders) 
     )
````

#### Task_3
3. Найдите топ-10 товаров по числу проданных единиц за всё время и постройте барплот.
    - Сколько было продано единиц товара с product_id==56?

````python
# Delete unnecessary column "Unnamed: 0"
df.drop(columns = 'Unnamed: 0', inplace= True) 
````

````python
# Redo old index to normal values (0...n-1)
df.reset_index(drop = True, inplace= True)
````

````python
df.head()
````

````python
top10_product_id = df\
                    .groupby('product_id', as_index = False)\
                    .agg({'quantity':'sum'})\
                    .sort_values('quantity', ascending= False) [:10]
top10_product_id
````

````python
query_quantity_product_id_56 = top10_product_id\
                            .query('product_id == 56')\
                            .quantity.values
query_quantity_product_id_56

print(
    '''Количество проданных единиц товара с ID № 56 составило - {} шт.'''.format(*list(query_quantity_product_id_56))
)
````

````python
# Барплот: топ-10 товаров по числу проданных единиц за всё время

plt.figure(figsize= (12,6))                                                             # задать масштаб графику
ax = sns.barplot(top10_product_id, x = 'product_id', y = 'quantity', palette= 'inferno', hue = 'quantity',  alpha = 0.75)   
# hue - оттенок по критерию, palette - задать цветовой паттерн изменения цвета (вне их значений), alpha - прозрачность (0-1)

ax.set(xlabel='Product ID', ylabel='Quantity')                                          # названия осей
sns.despine()                                                                           # убрать часть рамки графика
````

````python
# Переформатирование столбца с датой в формат "Даты"
df['date'] = pd.to_datetime(df['date'])
df.dtypes
````

#### Task_5
5. Сколько пользователей приобрели какой-либо товар повторно (более 1 раза)?

        Повтором будем считать покупку товара с одинаковым product_id, совершенную в разные дни.

````python
df.head()
````

````python
query_double_counts_customers = df\
    .groupby(['name', 'product_id'])\
    .agg({'date': pd.Series.nunique})\
    .rename(columns= {'date':'nunique_day_salles'})\
    .query('nunique_day_salles > 1')\
    .shape[0]

# 'date': pd.Series.nunique - количество уникальных дней в которые были совершены покупки по конкретным товарам #\
print('Количество пользователей, которые приобрели какой-либо товар повторно: {} чел.'.format(query_double_counts_customers))
````
