7) Небольшой перерыв на чистый питон. В списке numbers содержатся числа. Добавьте через цикл в список positive_numbers положительные числа из numbers. При этом, если вам попался 0, то нужно прекратить этот цикл.

Например, numbers = [1, -2, 3, 0, -3, 2] Тогда positive_numbers должен быть: positive_numbers = [1, 3]

```python
numbers = [1, -2, 3, -2, 4 ,0, -3, 2]

positive_numbers = []

scrininng = False

for i in numbers:
    if i == 0:
        scrininng = False
        break
        
    if i > 0:
        scrininng = True
        positive_numbers.append(i)
        continue
        
    if i < 0:
        scrininng = False

positive_numbers
```

## Ex_9

9) ⭐️Задание со звёздочкой!⭐️

Создайте колонку wait_time, в которой будет храниться разница между временем подачи машины (arrived_at) и временем заказа (start_at) в минутах. 

Чтобы сконвертировать полученный результат в минуты, можно использовать метод  .astype('timedelta64[m]')

```python
taxi['wait_time'] = (taxi.arrived_at - taxi.start_at) # добавлял модуль (abs), т.к. хз как может быть подача такси раньше чем время заказа
                        
taxi['wait_time']= taxi.wait_time.astype('timedelta64[m]')
taxi.head()
```

## Ex_9.1

9.1) ⭐️Задание со звёздочкой!⭐️

Посмотрим, какие водители опаздывали на заказы к определенному времени (start_type == 'reserved'). 

Опозданием считается время ожидания wait_time > 0.0.

Сгруппируйте отфильтрованные данные по driver_id и в качестве ответа укажите id водителя, который опоздал на заказ интересующего нас типа наибольшее число раз

```python
max_delays_driver = taxi\
                        .query("start_type == 'reserved' and wait_time > 0.0")\
                        .groupby('driver_id', as_index = False)\
                        .agg({'user_id':'count'})\
                        .rename(columns={'user_id':'count_delays'})\
                        .sort_values('count_delays', ascending = False) [:1]
max_delays_driver
```

```python
print(
'Водитель с id-номером {} опоздал на зарезервированный заказ больше всех - {} раз(а).'\
    .format(*list(max_delays_driver.driver_id), *list(max_delays_driver.count_delays))
)
```

```python
driver_delays = taxi\
                .query("start_type == 'reserved' and wait_time > 0.0")\
                .groupby('driver_id', as_index = False)\
                .agg({'user_id':'count'})\
                .rename(columns={'user_id':'count_delays'})\
                .sort_values('count_delays', ascending = False) 
driver_delays_info = driver_delays
driver_delays_info[['source', 'icon','driver_start_lat',  'driver_start_lon', 'start_at', 'end_lat', 'end_lon']] \
                    = taxi[['source', 'icon','driver_start_lat',  'driver_start_lon', 'start_at','end_lat', 'end_lon' ]]
driver_delays_info.head()
```

14) ⭐️Задание со звёздочкой!⭐️

Также мы можем посмотреть на несколько простых метрик, показывающих число уникальных пользователей за определённый период:

    DAU (daily active users) – число уникальных пользователей за день
    WAU (weekly active users) – число уникальных пользователей за неделю
    MAU (monthly active users) – число уникальных пользователей за месяц
    
Активные пользователи – те, кто воспользовался приложением и сделал хотя бы один заказ за выбранный промежуток времени. 

Постройте график MAU и выберите верные утверждения.


    Могут пригодиться: 
        - groupby – группировка nunique – число уникальных значений 
        - plot.line – линейная диаграмма

```python
taxi.head()
```

```python
#DAU (daily active users) – число уникальных пользователей за день

taxi['day_order'] = taxi.start_at.dt.date
DAU = taxi\
        .groupby('day_order')\
        .agg({'user_id':'nunique'})\
        .rename(columns={'user_id':'daily_active_users'})
DAU
```

```python
DAU.plot.line()
#DAU (daily active users) – число уникальных пользователей за день
```

```python
#WAU (weekly active users) – число уникальных пользователей за неделю
```

```python
taxi['weekofyear'] = taxi.start_at.dt.weekofyear

WAU = taxi\
        .groupby('weekofyear')\
        .agg({'user_id':'nunique'})\
        .rename(columns={'user_id':'weekly_active_users'})
WAU
```

```python
WAU.plot.line()
#WAU (weekly active users) – число уникальных пользователей за неделю
```

```python
# MAU (monthly active users) – число уникальных пользователей за месяц

MAU = taxi\
        .groupby(['month_order'])\
        .agg({'user_id':'nunique'})\
        .rename(columns={'user_id':'monthly_active_users'})
MAU
```

```python
MAU.plot.line()
# MAU (monthly active users) – число уникальных пользователей за месяц
```
