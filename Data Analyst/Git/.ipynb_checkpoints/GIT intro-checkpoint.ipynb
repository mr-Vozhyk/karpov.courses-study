### 1) Что делает команда cd?

__Перемещает по папкам, это очевидно!__

Открывает дисковод компьютера

Копирует файлы!

Выводит "Hello, World!"

Открыть дисковод!

**Сконкатенировать датафрэймы в один**



### 3) А команда pwd?

Сокращение от purge working directory - очистит её от незакомиченных файлов

**Вывести путь папки, где вы находитесь**

Ну теперь-то точно для открывания дисковода

Переместить файлы



### 4) Перейдём к более серьёзным вещам - что делает команда rm?

**Необратимо удаляет файл**

Печатает директорию, где мы находимся

Создаёт пустую директорию

Запускает терминал



### 5) А теперь создайте в вашей домашней директории несколько папок - test1, test2 и test3

Для создания нескольких папок можно перечислить их в одном вызове mkdir. 

Например

    mkdir d1 d2

Создаст папки d1 и d2



### 6) Внутри папки test1 создайте с помощью команды vi файл testing.py. 

Напишите в нём небольшой скрипт, где

    запринтите фразу Hi, I'm testing...
    
    выведите на печать каждое число из списка [10, 20, 30]
    
    выведите на печать фразу Finished 
    
Сохраните скрипт, убедитесь, что он находится в этой папке
